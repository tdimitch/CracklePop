```python
''' CracklePop 
    A program that prints out the numbers 1 to 100 (inclusive). 
    If the number is divisible by 3, it prints Crackle instead of the number. 
    If the number is divisible by 5, it prints Pop. 
    If the number is divisible by both 3 and 5, it prints CracklePop. '''

for i in range(100):
    if (i+1)%3 == 0:
        if (i+1)%5 == 0:
            print("CracklePop")
        else: 
            print("Crackle")
    elif (i+1)%5 == 0:
        print("Pop")
    else: 
        print(str(i+1))
```
